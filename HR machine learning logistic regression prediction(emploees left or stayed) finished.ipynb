```python
dummies1 = pd.get_dummies(data.salary)
dummies1
```

```python
dummies2 = pd.get_dummies(data.Department)
dummies2
```

```python
data_merged = pd.concat([dummies1, dummies2, data], axis='columns')
data_merged
```

```python
data_final = data_merged.drop(['Department', 'salary'], axis='columns')
data_final
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = data_final[['high', 'low', 'medium', 'IT', 'RandD', 'accounting', 'hr',
       'management', 'marketing', 'product_mng', 'sales', 'support',
       'technical', 'satisfaction_level', 'last_evaluation', 'number_project',
       'average_montly_hours', 'time_spend_company', 'Work_accident',
       'promotion_last_5years']]
```

```python
y = data_final['left']
```

```python
X.head()
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```
